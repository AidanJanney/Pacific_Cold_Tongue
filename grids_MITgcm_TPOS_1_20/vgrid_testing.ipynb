```python
import xarray as xr
import numpy as np
```

```python
cell_depth_z = xr.open_dataset('/glade/work/ajanney/Pacific_Cold_Tongue_Jofia/grids_MITgcm_TPOS_1_20/metrics_1500x480x185.nc')['Z']
cell_depth_z
```

```python
dz = np.abs(cell_depth_z.diff(dim='Z').to_numpy())
```

```python
from CrocoDash.vgrid import VGrid
vgrid = VGrid(dz=dz)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.close()

fig, ax = plt.subplots(1,2, figsize=(10,5))

# Create the plot
for depth in vgrid.z:
    ax[0].axhline(y=depth, linestyle='-', linewidth=0.5)  # Horizontal lines

ax[0].set_ylim(max(vgrid.z) + 10, min(vgrid.z) - 10)  # Invert y-axis so deeper values go down
ax[0].set_ylabel("Depth")

ax[1].plot(vgrid.dz,vgrid.z,'o-', markersize=2)
ax[1].set_ylim(max(vgrid.z) + 10, min(vgrid.z) - 10)

fig.suptitle("Vertical Grid")
plt.show()
```
